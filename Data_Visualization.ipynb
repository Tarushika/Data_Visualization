```python
df = pd.read_csv('/content/diabetes_data.csv')
df.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from mpl_toolkits.mplot3d import Axes3D

df_numeric = df.select_dtypes(include=np.number)

col1 = df_numeric.columns[0]
col2 = df_numeric.columns[1]
col3 = df_numeric.columns[2]

x = df_numeric[col1]

#Normal curves.
plt.figure(figsize=(6,4))
sns.histplot(x, bins=30, stat="density")
mu, std = stats.norm.fit(x)
xs = np.linspace(x.min(), x.max(), 100)
plt.plot(xs, stats.norm.pdf(xs, mu, std))
plt.show()

#Density and contour plots.
plt.figure(figsize=(6,4))
sns.kdeplot(x=df_numeric[col1], y=df_numeric[col2], fill=True)
plt.show()

#Correlation and scatter plots.
plt.figure(figsize=(6,4))
sns.scatterplot(x=df_numeric[col1], y=df_numeric[col2])
plt.show()

#Histograms.
plt.figure(figsize=(6,4))
plt.hist(df_numeric[col1], bins=30)
plt.show()

#Three dimensional plotting.
fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df_numeric[col1], df_numeric[col2], df_numeric[col3])
plt.show()
```
